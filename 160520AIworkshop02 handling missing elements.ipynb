```python
import pandas as pd
```

```python
titanicdata = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/Titanic_Data.csv')
```

```python
titanicdata.isnull()
```

```python
titanicdata.isnull().sum()
```

```python
titanicdata.drop(['Cabin'], axis=1, inplace= True)
```

```python
titanicdata['Age'].median()
```

```python
titanicdata['Age'].mean()
```

```python
import numpy as np
titanicdata['Age'].replace(np.nan, titanicdata['Age'].median(),inplace=True)
```

```python
titanicdata.isnull().sum()
```

```python
titanicdata.boxplot(figsize=(20,10))
```

```python
titanicdata[ 'Age'].plot(figsize = (20,8))
```

```python
diabetesdf = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/Diabetes_Preprocessing.csv')
```

```python
diabetesdf.head()
```

```python
diabetesdf.isnull().sum().sort_values(ascending=False)
```

```python
from sklearn.impute import SimpleImputer as SI

impute = SI(strategy='median')
diabetesdata = impute.fit_transform(diabetesdf)
```

```python
diabetesdata
```

```python
pd.DataFrame(diabetesdata, columns = diabetesdf.columns)
```
