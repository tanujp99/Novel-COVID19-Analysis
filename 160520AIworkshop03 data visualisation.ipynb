```python
import pandas as pd 

# matplotlib - easy graphs - bar, line, histogram, pie, scatter 
import matplotlib.pyplot as plt 

# seaborn - distribution - scatterplot, heatmap, density 
import seaborn as sns 

import numpy as np
```

```python
bostondf = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/boston_train.csv')
```

```python
bostondf.head()
```

```python
bostondf.isnull().sum()
```

```python
plt.figure(figsize= [20,10])
plt.plot(bostondf['AGE'])
plt.title('Line Graph', fontsize = 21)
plt.xlabel('Row Indexes', fontsize = 12)
plt.ylabel('Ages', fontsize = 12)
plt.show()
```

```python
bostondf['AGE'].sort_values().unique()
```

```python
pd.DataFrame(bostondf, columns = bostondf.columns)
```

```python
plt.figure(figsize=(20,10))
plt.hist(bostondf['MEDV'], label = 'Medv')
plt.title('Histogram', fontsize = 21)
plt.xlabel('MEDV', fontsize = 12)
plt.ylabel('Frequency', fontsize = 12)
plt.show()
```

```python
irisdf = pd.read_csv('https://raw.githubusercontent.com/CarmineCrown/MachineLearning/master/Datasets/iris.csv')
```

```python
irisdf.head()
```

### Scatter Plot

```python
plt.figure(figsize=(15,10))
sns.set_style('darkgrid')
plt.scatter(x = irisdf['sepal_length'], y = irisdf['sepal_width'], label = 'sepal_width') 
plt.scatter(x = irisdf['sepal_length'], y = irisdf['petal_length'], label = 'petal_length')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10)) 
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data = irisdf, hue = 'species')
plt.show()
```

### Density Grid

```python
plt.figure(figsize=(12,8)) 
sns.set_style('whitegrid') 
sns.distplot(bostondf['MEDV'], color = 'darkviolet', bins = 30) 
plt.show()
```
